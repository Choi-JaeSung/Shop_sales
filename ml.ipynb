## Preprocessing train_data

### Checking overlap

```python
train_data.duplicated().sum() # 중복된 데이터 x
```

### Checking nan

```python
train_data.isna().sum() # 6255 data 중에서 Promotion column에 4100개 이상의 nan값 존재
```

```python
train_data = train_data.fillna(0) # promotion nan값을 0으로 변경

train_data
```

### Encoding isHoliday

```python
train_data['IsHoliday'] = train_data.IsHoliday.map({True : 1, False : 0})

train_data
```

```python
train_data = train_data.drop(['Date'], axis=1)

X = train_data.iloc[:, :-1]
y = train_data.iloc[:, -1]
```

## Test data load

```python
test_data = pd.read_csv('./dataset/test.csv')

test_data.head(10)
```

```python
test_data.duplicated().sum()
```

```python
test_data.isna().sum()
```

```python
test_data = test_data.fillna(0)

test_data
```

```python
test_data.IsHoliday = test_data.IsHoliday.map({True : 1, False : 0})

test_data
```

```python
test_data = test_data.drop(['Date'], axis=1)

test_data
```

```python
X_test = test_data
```

### Train, Val split

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=2022, stratify=X.Store)

X_train.shape, X_val.shape
```

## XGBoost Learning

```python
import xgboost as xgb
```

```python
dtrain = xgb.DMatrix(data=X_train, label=y_train)
dval = xgb.DMatrix(data=X_val, label=y_val)
dtest = xgb.DMatrix(data=X_test)
```

```python
params ={'max_depth' : 3,
         'eta' : 0.2,
         'objective' : 'reg:linear',
         'eval_metric' : 'rmse',
         'lambda' : 1000
         }
```

```python
xgb_model = xgb.train(params=params, dtrain=dtrain, num_boost_round=2000, early_stopping_rounds=100, evals=[(dtrain, 'train'), (dval, 'eval')])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(xgb_model, params, refit=True)
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(dtest)
```
